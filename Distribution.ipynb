Ques : 1 Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use
BINOM.DIST)

- In a fair die, the probability of getting a 3 on one roll is 1/6. Since the die is rolled 30 times, the probability of getting exactly 5 threes can be calculated in Excel using the following formula

 =BINOM.DIST(5,30,1/6,FALSE)

 The result is approximately 0.176 or 17.6%. This means there is about a 17.6% chance of getting exactly five 3s in thirty rolls.

Ques : 3 A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 ?

- The values 45 and 55 are one standard deviation below and above the mean.

Mean = 50
Standard Deviation = 5

According to the Empirical Rule (68-95-99.7 Rule), about 68% of the observations in a normal distribution lie within one standard deviation of the mean.

Therefore, approximately 68% of the values lie between 45 and 55.

Ques : 5 What is Kurtosis and its types?
- Kurtosis is a statistical measure that shows how peaked or flat a distribution is compared to a normal distribution. It also indicates the presence of extreme values (outliers).

Types of Kurtosis:

Mesokurtic – Similar to a normal distribution (Kurtosis = 3).

Leptokurtic – Has a sharper peak and heavier tails, indicating more outliers.


Platykurtic – Has a flatter peak and lighter tails, indicating fewer outliers.

Kurtosis helps in understanding the shape and spread of data.

Ques : 8. How does log transformation help in stabilizing variance and making data more normally distributed?

- Log transformation is used when data is highly skewed or contains very large values. It reduces the effect of extreme values and makes the data more balanced.

The main benefits are:

Reduces right skewness.
Stabilizes variance.
Makes the distribution closer to normal.
Improves the performance of many statistical models.
Makes relationships between variables easier to analyze.

Because of these advantages, log transformation is widely used in statistics, finance, economics, and data analysis.
